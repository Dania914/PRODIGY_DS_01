```python

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Uploading file
uploaded = files.upload()

# Loading the dataset into pandas DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name, skiprows=4)

# Filtering the data for the year 2020
year = '2020'
filtered_data = df[['Country Name', year]].dropna()

# Sorting the data by population in 2020 for better visualization
filtered_data = filtered_data.sort_values(by=year, ascending=False)

# Plotting a bar chart for the distribution of the population in 2020
plt.figure(figsize=(12, 8))
plt.bar(filtered_data['Country Name'][:20], filtered_data[year][:20], color='orange')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Population Distribution in 2020 by Country')
plt.xticks(rotation=90)
plt.show()
```
